Define experiment configurations

```python
EXPERIMENT = 'dm_control'

# PUT IN THE PATH TO configs.py
cluster = TmuxCluster(
                     EXPERIMENT,
                     num_agents=2,
                     config_path='~/surreal/Surreal/surreal/main/ddpg_configs.py',
                     experiment_folder='/tmp/kurreal-experiments/',
                     start_dir='.',
                     preamble_cmds=['source activate surreal; export DISABLE_MUJOCO_RENDERING=1'], # or a list
                     config_command='--env "dm_control:cheetah-run"')
```
